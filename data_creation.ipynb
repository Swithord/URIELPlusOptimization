```python
from urielplus.urielplus import URIELPlus
import pandas as pd
from fancyimpute import SoftImpute
import numpy as np
```

```python
u = URIELPlus()
u.integrate_databases()
u.set_aggregation('U')
u.aggregate()
X = u.data[1][:].T
features = X[0].T
labels = u.feats[1]
langs = u.langs[1]
df = pd.DataFrame(features, columns=labels, index=langs)
df.to_csv('./data/URIELPlus_Union.csv')
```

```python
methods = ['mi']
for method in methods:
    for i in range(100, 701, 100):
        df = pd.read_csv(f'selection_result/base_{method}_{i}.csv', index_col=0)
        df_np = df.to_numpy()
        df_np = np.where(df_np == -1, np.nan, df_np)
        imputer = SoftImpute(max_iters=400,  max_value=1, min_value=0, init_fill_method="mean")
        imputed_values = imputer.fit_transform(df_np)
        df_imputed = pd.DataFrame(imputed_values, columns=df.columns, index=df.index)
        df_imputed.to_csv(f'selection_result/imputed_{method}_{i}.csv')
```
